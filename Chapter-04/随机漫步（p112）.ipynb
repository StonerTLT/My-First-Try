### 模拟一个随机漫步

#### 使用python自带random.randint模拟

```python
import random
```

```python
position = 0
walk = [position]
steps = 1000
for i in xrange(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)

x = xrange(steps+1)
y = walk
plt.plot(y)
# plt.plot(x, y)
# 也可以写成这种，在 x = range（len(y)）时，可以省略 x ，下同

plt.title('Random walk with +1/-1 steps by random')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

#### 使用numpy的random.randint模拟

```python
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum()

plt.title('Random walk with +1/-1 steps by Numpy.random')

plt.plot(walks)
plt.show()
```

```python
len(walks)
```

```python
walks.min()
```

```python
walks.max()
```

```python
(np.abs(walks) >= 10).argmax()
```

### 一次模拟多个随机漫步  
>#### (使用numpy.random.randint)

```python
nwalks = 5000
nsteps = 1000
draws =np.random.randint(0, 2, size=(nwalks,nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1)
```

```python
walks
```

```python
print walks.max(),walks.min()
```

```python
hits30 = (np.abs(walks) >= 30).any(1)
hits30
```

```python
# 到达30或者-30的数量
hits30.sum()
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)
```

```python
crossing_times
```

```python
crossing_times.mean()
```
